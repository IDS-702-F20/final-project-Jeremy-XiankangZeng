```python
import pandas as pd
```

```python
df = pd.read_csv("~/Desktop/archive/kills.csv")
```

```python
df.head()
```

```python
temp_list = df['Time'].to_list()
temp = []
for i in range(len(temp_list)):
    if temp_list[i] == -1:
        temp.append(0)
    else:
        temp.append(1)
    
```

```python
df['temp'] = temp
```

```python
df['temp'].value_counts()
```

```python
df['Time'].fillna(-1, inplace=True)
```

```python
df2 = df.groupby(['Address','Team'])['temp'].sum().to_frame().reset_index()
```

```python
bkills = df2[df2['Team'] == 'bKills']['temp'].to_list()
rkills = df2[df2['Team'] == 'rKills']['temp'].to_list()
diff = []
for i in range(len(bkills)):
    dif = bkills[i] - rkills[i]
    diff.append(dif)
```

```python
address = list(df2['Address'].unique())
```

```python
dic = {}
for i in range(len(address)):
    dic[address[i]] = diff[i]
```

```python
list(dic.keys())
```

```python
diff_frame = pd.DataFrame({'Address' : list(dic.keys()) , 'diff' : list(dic.values())})
```

```python
main = pd.read_csv('/Users/jer/Desktop/archive/LeagueofLegends.csv')
```

```python
main = main.merge(diff_frame, left_on = 'Address', right_on = 'Address', how = 'left')
```

```python
main.to_csv('main.csv')
```
